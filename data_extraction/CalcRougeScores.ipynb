```
import pandas as pd

from rouge import Rouge 

import nltk
from nltk.tokenize import sent_tokenize

nltk.download('punkt')
```

```
articles_df = pd.read_csv('drive/My Drive/!UBS/DW/covid_main.csv', index_col=0, dtype ={'body': str})
print(articles_df.shape)
articles_df.head()
```

### Score each sentence of body against all sentences of abstract and select best score

```
rouge = Rouge()
shas = []
sentences = []
scores = []

for row in articles_df.iloc:
    if type(row['body']) != str:  # Handle strange type error
        continue

    sha = row['sha']
    body_sents = [sent for sent in sent_tokenize(row['body']) if 10 < len(sent) < 150] # Filter thrash
    abstract_sents = [sent for sent in sent_tokenize(row['abstract']) if 10 < len(sent) < 150] # Filter thrash

    best_score = 0
    for body_sent in body_sents:
        for abstract_sent in abstract_sents:
            score = rouge.get_scores(body_sent, abstract_sent)[0]['rouge-2']['f']
            if score > best_score:
                best_score = score
        shas.append(sha)
        sentences.append(body_sent)
        scores.append(best_score)
```

```
sentences_df = pd.DataFrame({'sha': shas, 'sentence': sentences, 'score': scores})

print(sentences_df.shape)
sentences_df.head()
```
