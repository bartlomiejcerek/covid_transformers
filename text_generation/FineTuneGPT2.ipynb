```
import pandas as pd
```

```
articles_df = pd.read_csv('/content/drive/My Drive/!UBS/DW/covid.csv', index_col=0)
articles_df.head()
```

## Dump all text to TXT file.

```
with open("text_dump.txt","w") as f:
    for row in articles_df.iloc:
        for el in ['title','authors','abstract','introduction','body']:
            f.write(str(row[el]) + '\n\n')
```
