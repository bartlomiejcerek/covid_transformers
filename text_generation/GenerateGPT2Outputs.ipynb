```
import transformers
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

import pandas as pd 
import numpy as np 

import os
import json
```

```
# add the EOS token as PAD token to avoid warnings
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model_org = GPT2LMHeadModel.from_pretrained("gpt2", pad_token_id=tokenizer.eos_token_id)
model_tuned = GPT2LMHeadModel.from_pretrained('drive/My Drive/!UBS/DW/gpt2_model', pad_token_id=tokenizer.eos_token_id)
model_org.cuda()
model_tuned.cuda();
```

```
# Function for generating text using top k and top p sampling
def generate_text(model, tokenizer, input_text, length):
    sample_outputs = model.generate(
        tokenizer.encode(input_text, return_tensors='pt').cuda(),
        do_sample=True, 
        max_length=length, 
        top_k=40, 
        top_p=0.95)
        
    return tokenizer.decode(sample_outputs[0])
```

## Generate Text

```
for i in range(5):
    torch.manual_seed(i)
    input_text = 'Symptomes of the virus are'
    length = 100

    print(20*'-', 'ORIGINAL MODEL', 20*'-')
    print(generate_text(model_org, tokenizer, input_text, length))
    print(20*'-', 'TUNED MODEL', 20*'-')
    print(generate_text(model_tuned, tokenizer, input_text, length))
    print('\n\n')
```
